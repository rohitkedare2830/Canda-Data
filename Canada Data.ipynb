# Convert Budget to Numeric

```python
df['budget'] = df['budget'].astype(float)
```

```python
df
```

```python
df
```

```python
df['project_id'] = pd.util.hash_pandas_object(df['title']).astype(str)
```

```python
df
```

```python
df['project_name'] = df['title'].str.strip().str.title()
```

```python
df
```

```python
df['budget_usd_clean'] = df['ml.budget_ccy_bot.results.budget_usd']

mask = df['budget_usd_clean'].isna() & (df['currency']=='USD')
df.loc[mask, 'budget_usd_clean'] = df.loc[mask, 'budget']
```

```python
df
```

```python
import ast

def extract_start(x):
    try:
        d = ast.literal_eval(x)
        return d.get('project_start_date', None)
    except:
        return None

df['timestamp_clean'] = df['timestamps'].apply(extract_start)
```

```python
df
```

```python
df = df[~df['title'].str.contains("DUPLICATE", na=False)]
```

```python
df
```

```python
infra_keywords = [
    "road", "bridge", "metro", "rail", "airport", "port",
    "water", "hospital", "school", "power", "solar",
    "plant", "pipeline", "dam", "desalination"
]

df['is_infrastructure_project'] = df['description'].str.lower().apply(
    lambda x: any(word in str(x) for word in infra_keywords)
)
df
```

```python
df['is_megaproject'] = df['budget_usd_clean'] >= 500_000_000
df
```

```python
failure_status = ['Failure','Abandoned','Canceled','Delayed']

df['failed_or_problematic'] = df['ml.status_simplified.results.identified_status'].isin(failure_status)
df
```

```python
analysis = df.groupby('ml.procurement_method_redo2.results.method')['failed_or_problematic'].value_counts(normalize=True)*100
analysis
df
```

```python
failure_rate = df.groupby('ml.procurement_method_redo2.results.method')['failed_or_problematic'].mean()*100

plt.figure()
failure_rate.plot(kind='bar')
plt.title("Failure Rate by Delivery Method")
plt.xlabel("Delivery Method")
plt.ylabel("Failure %")
plt.show()
```
